```python
import pandas as pd
import yfinance as yf
import plotly.express as px
from statistics import NormalDist
```

```python
SPY =yf.download("^gspc", start="2000-01-01", end="2021-05-18")
SPY['Adj Close'] = SPY['Adj Close'].pct_change()*100
SPY.reset_index(inplace=True)
SPY['Date_str'] = SPY['Date'].astype(str)
fig = px.strip(SPY.reset_index(), x = 'Adj Close', orientation='h',
labels={"variable":" ", "value":"Percentage Change"},
title="Daily Percentage Change of SPY", hover_data=['Adj Close', 'Date_str'], template="ggplot2")
```

```python
fig.update_traces(jitter = 0.5).update_yaxes(range=[-0.13,0.13]).update_xaxes(nticks=16).show()
```

```python
import numpy as np

perc_change = SPY["Adj Close"]
```

```python
user_move = float(input("What is the move?"))

if user_move > 0: 
    prob = 1 - NormalDist(sigma=np.std(perc_change), mu=np.mean(perc_change)).cdf(user_move)
    display = f"The probability it moves {user_move}% is {round(prob,3)}"

elif user_move < 0: 
    prob = NormalDist(sigma=np.std(perc_change), mu=np.mean(perc_change)).cdf(user_move)
    display = f"The probability it moves {user_move}% is {round(prob,3)}"
```

```python
import plotly.figure_factory as ff 

x1 = np.random.normal(np.mean(perc_change)*100, np.std(perc_change)*100, 500)
hist_data = [x1]
group_labs = ['Probability']

ff.create_distplot(hist_data, group_labs)
```
